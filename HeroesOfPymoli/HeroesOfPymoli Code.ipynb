## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#remove repeating values in series 'SN'
unique_purchase_data = purchase_data.drop_duplicates(subset="SN",keep='first',inplace=False)
#create a list for bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
#create a list for group labels
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#add Age Group column using pd.cut
unique_purchase_data["Age Group"] = pd.cut(unique_purchase_data["Age"], bins, labels=group_labels)
#group by Age Group, use .count to count values 
age_data = unique_purchase_data.groupby("Age Group").count()
#drop columns we don't need
age_data = age_data.drop(['Purchase ID', 'SN', 'Gender', "Item ID", 'Item Name', 'Price'], axis=1)
# calculate percentage
percentage = (age_data['Age']/age_data['Age'].sum()) * 100
#remane column 'Age' to 'Total Count'
age_data = age_data.rename(columns={'Age': 'Total Count'})
#add column for 'Percentage of Players'
age_data['Percentage of Players'] = percentage
#format percentage
age_data["Percentage of Players"] = age_data["Percentage of Players"].map("{:.2f}%".format)
#create new dataframe
age_data = pd.DataFrame(age_data)

age_data
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# File to Load (Remember to Change These) - data frame was altered when we used .drop so we need to reload the .csv
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
#create a list for bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
#create a list for group labels
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#add Age Group column using pd.cut
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_labels, include_lowest=True)
#group by Age Group
purchase_age_data = purchase_data.groupby("Age Group")
#count 'SN' to calculate purchase count
purchase_count = purchase_age_data["SN"].count()
#find average price
average_purchase = purchase_age_data["Price"].mean()
#find total purchase value
total_purchase = purchase_age_data["Price"].sum()
#find average price per person
average_per_person = (total_purchase/purchase_count)

#create new dataframe
purchase_age_df = pd.DataFrame({"Purchase Count": purchase_count,
                                 "Average Purchase Price": average_purchase,
                                 "Total Purchase Value": total_purchase,
                                 "Average Purchase Total per Person": average_per_person})
#format to show as currency
purchase_age_df["Average Purchase Price"] = purchase_age_df["Average Purchase Price"].map("${:.2f}".format)
purchase_age_df["Total Purchase Value"] = purchase_age_df["Total Purchase Value"].map("${:.2f}".format)
purchase_age_df["Average Purchase Total per Person"] = purchase_age_df["Average Purchase Total per Person"].map("${:.2f}".format)

purchase_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group by SN
top_spender_data = purchase_data.groupby("SN")
#count purchase per SN
top_purchase_count = top_spender_data["Purchase ID"].count()
#find average price
top_avg_price = top_spender_data["Price"].mean()
#find total purchase value
top_total_price = top_spender_data["Price"].sum()
#create new dataframe
top_spender_df = pd.DataFrame({"Purchase Count": top_purchase_count,
                               "Average Purchase Price": top_avg_price,
                               "Total Purchase Value": top_total_price,})
#sort from high to low
top_spender_df = top_spender_df.sort_values(by='Total Purchase Value', ascending=False)
#format Average Purchase Price and Total Purchase Value as currecny 
top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].map("${:.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:.2f}".format)
#show preview of dataframe
top_spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group by Item ID and Item Name
pop_item_data = purchase_data.groupby(["Item ID", "Item Name"])
#count number of times item has been purchased
pop_purchase_count = pop_item_data["Purchase ID"].count()
#calculate Total Purchase Value
pop_total_price = pop_item_data["Price"].sum()
#calculate price of item
pop_price = pop_total_price/pop_purchase_count
#create new dataframe
pop_item_df = pd.DataFrame({"Purchase Count": pop_purchase_count,
                               "Item Price": pop_price,
                               "Total Purchase Value": pop_total_price,})
#sort high to low
pop_item_df = pop_item_df.sort_values(by='Purchase Count', ascending=False)
#format Item Price and Total Purchase Value as currecny 
pop_item_df["Item Price"] = pop_item_df["Item Price"].map("${:.2f}".format)
pop_item_df["Total Purchase Value"] = pop_item_df["Total Purchase Value"].map("${:.2f}".format)

pop_item_df.head()
```
